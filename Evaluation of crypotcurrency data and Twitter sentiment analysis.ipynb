### Importing necessary libraries and reading datasets 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 



bitcoin = "coin_Bitcoin.csv"
eth = "coin_Ethereum.csv"
doge = "coin_Dogecoin.csv"
```

### BITCOIN ANALYSIS

```python
df_bitcoin = pd.read_csv(bitcoin)
plt.figure(figsize=(20,20))
fig, ax = plt.subplots()
 
# Plotting the closing price chart for entire timeline 
plt.plot(df_bitcoin['Date'], df_bitcoin['Close'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
fig.autofmt_xdate()
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Price($)")
plt.title("Bitcoin Chart")
plt.savefig(fname="Bitcoin.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Plotting Candlestick chart using plotly 


import plotly.offline as py
import plotly.graph_objs as go

py.init_notebook_mode(connected=True)

data_bit = [go.Candlestick(x=df_bitcoin["Date"],
                       open=df_bitcoin["Open"],
                       high=df_bitcoin["High"],
                       low=df_bitcoin["Low"],
                       close=df_bitcoin["Close"])]
layout_bit = go.Layout(title='Bitcoin Candlestick',
                   xaxis={'rangeslider':{'visible':True}})
fig = go.Figure(data=data_bit,layout=layout_bit)
fig.update_layout(yaxis_title="Price($)")
py.iplot(fig,filename='Bitcoin_candlestick')
```

```python
# Plotting Volume bar plot for specific time interval 

#df_bitcoin['Date'] = pd.to_datetime(df_bitcoin['Date'], format='%Y-%m-%d')
start_date = "2021-01-25"
end_date = "2021-02-02"
after_start_date = df_bitcoin["Date"] >= start_date
before_end_date = df_bitcoin["Date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = df_bitcoin.loc[between_two_dates]
filtered_dates

plt.yscale('log')
barplot = sns.barplot(filtered_dates['Date'], filtered_dates['Volume'])
plt.xticks(rotation=90)
plt.ylabel('Volume traded in $ ')
plt.title("Bitcoin Volume Chart")
plt.savefig(fname="bitcoin volume.jpg", dpi=300, bbox_inches='tight')
```

```python
# plotting closing price chart for specific time interval 

plt.figure(figsize=(20,20))
fig, ax = plt.subplots()
 

plt.plot(filtered_dates['Date'], filtered_dates['Close'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
fig.autofmt_xdate()
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Price($)")
plt.title("Bitcoin Chart")
plt.savefig(fname="Bitcoin.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

### Scraping Twitter data to collect Elon Musk's Tweets for a specific time interval  


```python
import snscrape.modules.twitter as sntwr
import pandas as pd 

#creating a list to append tweets
tweetslst = []

#using twittersearchscraper to scrape data and append tweets to list 
for i, tweet in enumerate(sntwr.TwitterSearchScraper('from:elonmusk since:2020-12-20 until:2021-01-30').get_items()):
    if i > 10000:
        break
    tweetslst.append([tweet.date, tweet.id, tweet.content])
    
#creating a dataframe 

tweets_df_elon = pd.DataFrame(tweetslst, columns=['Datetime', 'Tweet_id', 'text'])
```

##### Saving collected tweets in a csv file 

```python


tweets_df_elon 
tweets_df_elon.to_csv('Elon_Tweets.csv')
```

```python
tweets = pd.read_csv('Elon_Tweets.csv')
tweets
```

```python
# defining a function to clean all the @, # , whitespaces and hyperlinks using regex

import re


def cleanText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text)
    text = re.sub(r'#', '', text)
    # text = re.sub(r'RT[\s]+', '', text)  ----> to remove RT's with whitespaces
    text = re.sub(r'http\S+', '', text) # ----> to remove hyperlinks
    
    return text

tweets['text'] = tweets['text'].apply(cleanText)
tweets
```

```python
from wordcloud import WordCloud
from textblob import TextBlob 


# getting the word cloud 

allwords = ' '.join([x for x in tweets['text']])
wordcloud = WordCloud(width = 500, height=300, max_words=100, background_color='black', random_state=15 ).generate(allwords)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# Sentiment Analysis getting polarity and subjectivity 

# Defining functions to get subjectivity for the tweets 

def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity


# Defining functions to get polarity for the tweets 
def getPolarity(text):
    return TextBlob(text).sentiment.polarity 
```

```python
# Creating new columns for polarity and subjectivity

tweets['Subjectivity'] = tweets['text'].apply(getSubjectivity)
tweets['Polarity'] = tweets['text'].apply(getPolarity)

tweets
```

```python
# Plotting the polarity and subjectivity


plt.scatter(tweets['Polarity'], tweets['Subjectivity'], color='blue')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.title("Sentimental Analysis")
plt.show()
```
